```
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```
df = pd.read_csv("/content/books (1).csv")
df.head()
```

```
df['combined_features'] = (
    df['Book-Title'].fillna('') + ' ' +
    df['Book-Author'].fillna('') + ' ' +
    df['Publisher'].fillna('')
)
```

```
vectorizer = TfidfVectorizer(stop_words='english')
tfidf_matrix = vectorizer.fit_transform(df['combined_features'])
```

```
similarity_matrix = cosine_similarity(tfidf_matrix)
```

```
def recommend_books(book_title, df, similarity_matrix, top_n=10):
    if book_title not in df['Book-Title'].values:
        return "Book not found in database."

    index = df[df['Book-Title'] == book_title].index[0]
    similarity_scores = list(enumerate(similarity_matrix[index]))
    sorted_scores = sorted(similarity_scores, key=lambda x: x[1], reverse=True)[1:top_n+1]

    recommended_books = []
    for i, score in sorted_scores:
        recommended_books.append({
            "Book-Title": df.iloc[i]['Book-Title'],
            "Book-Author": df.iloc[i]['Book-Author'],
            "Publisher": df.iloc[i]['Publisher'],
            "Similarity Score": round(score, 3)
        })

    return recommended_books
```

```
recommend_books("Decision in Normandy", df, similarity_matrix, top_n=5)
```
